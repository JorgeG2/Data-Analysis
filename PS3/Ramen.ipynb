```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### 1. Read the data from the CSV file into a DataFrame and display the first five rows.

```python
my_file = 'ramen-ratings.csv'
dp = pd.read_csv(my_file)
```

```python
dp.head()
```

### 2. Create a bar plot to show the Stars by Style with a Seaborn specific method.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Style', y='Stars', data=dp, ci=None, palette='viridis')

# plt.title('Average Stars by Style', fontsize=15)
# plt.xlabel('Style', fontsize=12)
# plt.ylabel('Average Stars', fontsize=12)
plt.xticks(rotation=45)  # Rotate labels to improve readability

plt.show()
```

### 3. Create a bar plot to show the Stars by Style with a Seaborn general method.

```python
sns.catplot(x='Style', y='Stars', data=dp, kind='bar', ci=None, palette='viridis', height=6, aspect=(10/6))

plt.xticks(rotation=45)  # This might not be directly applicable as sns.catplot handles facets

plt.show()
```

### 4. Modify the plot you just created to use a custom title and x-axis label.

```python
g = sns.catplot(x='Style', y='Stars', data=dp, kind='bar', ci=None, palette='viridis', height=6, aspect=(10/6))

# Set the title with a higher level of customization
g.fig.suptitle('Custom Title: Average Stars by Style', fontsize=15, y=1.05)

# Since sns.catplot() returns a FacetGrid, setting x-axis label for all subplots
g.set_axis_labels('Custom X-Axis Label: Ramen Style', 'Average Stars')

# To rotate the x-ticks, you operate on the axes of the FacetGrid
g.set_xticklabels(rotation=45)

plt.show()
```

```python
## 5. Use the query() method to filter the data to contain only data for Japan, India, Taiwan, and the U.S. Plot the stars by style for each country on a subplot with two subplots per row.
```

```python
# Filter the DataFrame to include only the specified countries
filtered_dp = dp[dp['Country'].isin(['Japan', 'India', 'Taiwan', 'USA'])]

# Creating a bar plot that compares the Stars by Style across the specified countries
plt.figure(figsize=(12, 6))
sns.barplot(data=filtered_dp, x='Style', y='Stars', hue='Country', ci=None, palette='viridis')

plt.title('Comparison of Ramen Stars by Style in Japan, India, Taiwan, and the USA', fontsize=14)
plt.xlabel('Ramen Style', fontsize=12)
plt.ylabel('Average Stars', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Country')

plt.tight_layout()
plt.show()
```
