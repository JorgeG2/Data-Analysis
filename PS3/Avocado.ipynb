```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### 1. Read the data from the CSV file into a DataFrame and display the first five rows.

```python
my_file = 'avocado.csv'
dp = pd.read_csv(my_file)
```

```python
dp.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dp, x='AveragePrice', y='Total Volume')

# Set the title and labels
plt.title('Scatter Plot of Average Price vs Total Volume', fontsize=14)
plt.xlabel('Average Price', fontsize=12)
plt.ylabel('Total Volume', fontsize=12)

# Show the plot
plt.show()
```

### 3. Create the same plot with a general method.

```python
# Create the scatter plot using Matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(dp['AveragePrice'], dp['Total Volume'])

# Set the title and labels
plt.title('Scatter Plot of Average Price vs Total Volume', fontsize=14)
plt.xlabel('Average Price', fontsize=12)
plt.ylabel('Total Volume', fontsize=12)

# Show the plot
plt.show()
```

### 4. Create the same plot but with the hue parameter set to year and the dots for the total volume ranging in size from 10 to 100.

```python
size_range = (10, 100)

# Create the scatter plot with Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dp.sample(frac=0.1), x='AveragePrice', y='Total Volume', hue='year', size='Total Volume', sizes=size_range)

# Set the title and labels
plt.title('Scatter Plot of Average Price vs Total Volume', fontsize=14)
plt.xlabel('Average Price', fontsize=12)
plt.ylabel('Total Volume', fontsize=12)

# Show the plot
plt.show()
```
