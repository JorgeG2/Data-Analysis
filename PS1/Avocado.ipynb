### 3. Which region has the lowest average price for organic avocados? Hint: Create wide data from the grouped data that you created in task 8

```python
organic_df = df[df['type'] == 'organic']
organic_df 
```

```python
# Group by region and calculate the average price
average_price_organic = organic_df.groupby('region')['AveragePrice'].mean()

# Find the region with the lowest average price
lowest_price_region = average_price_organic.idxmin()

# Output the region and its lowest average price
lowest_price_region, average_price_organic[lowest_price_region]
```
